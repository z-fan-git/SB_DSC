## 3 Adopted User Determination<a id='3'></a>

* Defining  an  "adopted  user"   as  a  user  who   has  logged  into  the  product  on  three  separate days  in  at  least  one  seven day  period.

```python
user_engagement.head()
```

```python
user_engagement['date'] = user_engagement['time_stamp'].dt.strftime('%Y-%m-%d')
```

```python
user_engagement.head(10)
```

```python
records = pd.pivot_table(user_engagement, values='visited', 
                                index='date', 
                                columns='user_id', 
                                aggfunc=np.sum)
records.head(10)
```

```python
adopted_user_id = []
for i in range(records.shape[1]):
    if sum(records.iloc[:,i].rolling(7).count() > 3) > 0:
        adopted_user_id.append(i)
```

```python
len(adopted_user_id)
```

```python
user['adopted'] = 0
user.loc[user['object_id'].isin(adopted_user_id), 'adopted'] = 1
```

```python
user.head()
```

```python
print(f'The ratio of adopted users is {round(user.adopted.sum()/user.object_id.nunique()*100, 2)} %')
```

```python
from sklearn.model_selection import train_test_split, cross_validate, cross_val_score, GridSearchCV
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

**Train/Test Split**

Create X, y variables for train/test split. 

```python
data = user.copy()
# customer info is removed as they are not related to targeting feature
data = data.drop(columns=['name', 'email', 'object_id','invited_by_user_id'])
```

```python
# Convert datetime to numerical features
data['creation_time'] = data['creation_time'].dt.year*10000+data['creation_time'].dt.month*100+data['creation_time'].dt.day
data['last_session_creation_time'] = data['last_session_creation_time'].dt.year*10000+data['last_session_creation_time'].dt.month*100+data['last_session_creation_time'].dt.day
```

```python
data.info()
```

We can't make up missing data for last_session_creation_time and invited_by_user_id and fill missing values with other information. We need to remove all missing records.

```python
# fill the missing part with median
data['last_session_creation_time'] = data['last_session_creation_time'].fillna(data['last_session_creation_time'].median())
```

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

```python
X = data.drop(columns = ['adopted'])
y = data['adopted']
print(X.shape,'\n', y.shape)
```

We will split the data with test_size = 0.25, i.e., training uses 75% of data.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=47)
print(X_train.shape,'\n', X_test.shape)
```

```python
X_train.info()
```

**Modeling**

Random forest classifer is applied to predict the retention because:
- it is robust to outliers, scalable, and able to naturally model non-linear decision boundaries thanks to their hierarchical structure.
- Data Scaling is not necessary for Ramdorm Forest
- it provide the importance of features, which is critical to figuring out bussiness strategies.

```python
classifiers = RandomForestClassifier(random_state=42)
classifiers.fit(X_train, y_train)
y_pred = classifiers.predict(X_test)

precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
score = classifiers.score(X_test, y_test)
```

```python
feature_importance = pd.DataFrame(classifiers.feature_names_in_, columns=['feature'])
feature_importance['importance'] = classifiers.feature_importances_
feature_importance.sort_values('importance', ascending = False)
```

**Key factors predict future user adoption**
* org_id:   the  organization  (group  of  users)  they  belong  to
* creation_time:  when  they  created  their  account
* last_session_creation_time:   unix  timestamp  of  last  login
